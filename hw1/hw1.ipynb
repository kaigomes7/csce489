Imports `pandas` library to load data,`numpy` for built-in math functions, and `matplotlib` to plot the histograms

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load data from file into a pandas DataFrame
* `pd.read_fwf` reads in the data from the .data file
* `na_values` tells pandas what to treat as NaN in the DataFrame
* `names` applies the names to the columns

```python
data = pd.read_fwf("auto-mpg.data", na_values='?', names=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name'])
```

Define `print_stats` function to print the stats for the desired columns of the dataset

```python
def print_stats(series, units):
    print(f"Units: {units}")
    print(f"Min {series.name}: {np.min(series)}")
    print(f"Max {series.name}: {np.max(series)}")
    print(f"Mean {series.name}: {np.mean(series)}")
    print(f"Std {series.name}: {np.std(series)}")
    print("------------------------------------")
```

Call `print_stats` function for mpg, displacement, horsepower, and weight to display statistics for each column

```python

print_stats(data['mpg'], 'miles/gallon')
print_stats(data['displacement'], 'cu (cubic inches)')
print_stats(data['horsepower'], 'hp')
print_stats(data['weight'], 'lbs')
```

Define `generate_histogram` function that create histograms of mpg, acceleration, horsepower and weight based on the number of cyliners passed in

```python
def generate_histogram(cylinders):
    subset = data[data['cylinders'] == cylinders]
    mpg = subset['mpg']
    acc = subset['acceleration']
    hrsprs = subset[subset['horsepower'].notna()]['horsepower']
    weight = subset['weight']
    fig, axs = plt.subplots(2, 2)
    axs[0, 0].hist(mpg)
    axs[0, 0].set_title('MPG')
    axs[0, 0].set(xlabel='mpg', ylabel='# of cars')
    axs[0, 1].hist(acc)
    axs[0, 1].set_title('Acceleration')
    axs[0, 1].set(xlabel='acceleration [0-60 mph] (seconds)', ylabel='# of cars')
    axs[1, 0].hist(hrsprs)
    axs[1, 0].set_title('Horsepower')
    axs[1, 0].set(xlabel='horsepower', ylabel='# of cars')
    axs[1, 1].hist(weight)
    axs[1, 1].set_title('Weight')
    axs[1, 1].set(xlabel='weight (lbs)', ylabel='# of cars')
    fig.suptitle(f"{cylinders} Cylinder Histograms")
    fig.tight_layout()
```

Display histograms for cars with 4 cylinders

```python
generate_histogram(4)
```

Display histograms for cars with 6 cylinders

```python
generate_histogram(6)
```

Display histograms for cars with 8 cylinders

```python
generate_histogram(8)
```
